```python
import pandas as pd

# Baixar um arquivo de dados pequeno (amostra pública do ENEM)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

dados = pd.read_csv(url)

# Mostrar as primeiras linhas
dados.head()
```

```python
# Quantas linhas e colunas tem o conjunto de dados?
dados.shape
```

```python
# Média do valor da conta
media = dados["total_bill"].mean()
media
```

```python
import matplotlib.pyplot as plt

# Criar gráfico simples
dados["total_bill"].hist()

plt.title("Distribuição do valor da conta")
plt.xlabel("Valor da conta")
plt.ylabel("Frequência")

plt.show()
```
